```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# 1. 정찰 및 감시 임무 사례 표 데이터 생성
recon_table = pd.DataFrame({
    "국가/운영체제": ["미국", "터키", "이스라엘", "러시아"],
    "드론봇 종류": ["MQ-9 리퍼 (Reaper)", "바이락타르 TB2", "헤론 TP", "오리온 (Orion)"],
    "임무 유형": ["적군 동선 추적", "실시간 정찰 및 감시", "대규모 지역 정찰", "실시간 정보 수집"],
    "주요 성과": [
        "테러리스트 은신처 식별 및 타격 지원",
        "아르메니아 방공망 탐지 및 무력화",
        "국경 지역 침입 탐지 및 보고",
        "병력 이동 경로 추적 및 기습 공격 대응"
    ],
    "출처": ["(Kim & Park, 2021)", "(Yu, 2021)", "(Choi, 2020)", "(Lee & Kim, 2019)"]
})

# 2. 다중 드론 협력 시스템 그림 데이터 생성
drone_positions = {
    "Drone A": (0, 1),
    "Drone B": (1, 0),
    "Drone C": (0, -1),
    "Target": (2, 0)
}

edges = [("Drone A", "Drone B"), ("Drone B", "Drone C"), ("Drone B", "Target")]

# 3. 방어 체계 통합 사례 표 데이터 생성
defense_table = pd.DataFrame({
    "국가/운영체제": ["이스라엘", "러시아", "미국"],
    "드론봇 종류": ["아이언 돔", "라자레프 (Lazarev)", "C-UAS 시스템"],
    "방어 임무": ["적 로켓 요격 및 차단", "신호 교란 및 방향 전환", "드론 탐지 및 요격"],
    "주요 성과": [
        "공격 로켓 및 드론 90% 이상 차단",
        "적 드론 70% 이상 무력화",
        "대규모 드론 공격 방어"
    ],
    "출처": ["(Yu, 2021)", "(Park, 2020)", "(Kim & Lee, 2021)"]
})

# Plotting and saving visualizations
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('tight')
ax.axis('off')
table_recon = ax.table(cellText=recon_table.values, colLabels=recon_table.columns, loc="center", cellLoc='center')
plt.title("정찰 및 감시 임무 사례 분석")
plt.savefig('/mnt/data/recon_table.png', bbox_inches='tight')
plt.close()

# Drawing network for multi-drone cooperation system
fig, ax = plt.subplots(figsize=(8, 6))
G = nx.DiGraph()
G.add_edges_from(edges)
nx.draw(
    G, pos=drone_positions, with_labels=True, node_size=3000, node_color="lightblue", font_size=10,
    edge_color="gray", font_weight="bold", arrowsize=20, ax=ax
)
plt.title("다중 드론 협력 시스템의 작전 사례")
plt.savefig('/mnt/data/multi_drone_system.png', bbox_inches='tight')
plt.close()

fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('tight')
ax.axis('off')
table_defense = ax.table(cellText=defense_table.values, colLabels=defense_table.columns, loc="center", cellLoc='center')
plt.title("방어 체계 통합 사례")
plt.savefig('/mnt/data/defense_table.png', bbox_inches='tight')
plt.close()

"/mnt/data/recon_table.png", "/mnt/data/multi_drone_system.png", "/mnt/data/defense_table.png"
```
